```python
movies = pd.read_table('hetrec2011-movielens-2k-v2/movies.dat', sep='\t', encoding='latin1', na_values=['\\N'])
movies.head()
```

```python
ratings = pd.read_table('hetrec2011-movielens-2k-v2/user_ratedmovies-timestamps.dat', sep='\t')
ratings.head()
```

```python
movie_genres = pd.read_table('hetrec2011-movielens-2k-v2/movie_genres.dat', sep='\t')
movie_genres.head()
```
