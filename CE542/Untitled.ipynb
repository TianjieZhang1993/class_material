```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('ce540.xlsx',sheet_name='Sheet2')
data['test4-ys'].max()
```

```python
plt.plot(data['test2-ys'])
plt.hlines(y=data['test2-ys'].max(), color='r', xmin = 0, xmax = 100,linestyle='-')
plt.annotate("Static yield stress", xy=(60, 50), xytext=(120, 40),arrowprops=dict(arrowstyle="->"))

plt.hlines(y=data['test2-ys'][300], color='r', xmin = 200, xmax = 350,linestyle='--')
plt.annotate("Dynamic yield stress", xy=(270, 23), xytext=(120, 10),arrowprops=dict(arrowstyle="->"))
# plt.title('yield stress')
plt.xlabel('Time(s)')
plt.ylabel('Shear stress(Pa)')
plt.tight_layout()
plt.savefig('static_yield_stress.png',dpi=600,bbox_inches = 'tight')
```
