```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from census import Census
from us import states
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
rng = np.random.RandomState(20211028)
```

```python
haberman=pd.read_csv('haberman.data')
haberman.head()
```

```python
haberman.rename(columns={'30':'Age',1.1": "Survival status",})
```

```python
test = haberman.sample(frac=0.2, random_state=rng)
```

```python
train_mask = pd.Series(True, index=haberman.index)
train_mask[test.index] = False
train = haberman[train_mask].copy()
train.head()
```

```python
mod = smf.glm('Survival status ~ 30 + 64 + 1', train,
              family=sm.families.Binomial())
lg_res = mod.fit()
lg_res.summary()
```
