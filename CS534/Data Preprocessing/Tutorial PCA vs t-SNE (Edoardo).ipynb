```python
import matplotlib.pyplot as plt
def color(s):
    if s=='Iris-virginica':
        return 'red'
    else:
        if s=='Iris-versicolor':
            return 'blue'
    return 'yellow'


import numpy as np
import pandas as pd
from sklearn import preprocessing
aa=pd.read_csv("iris.csv")
d=aa.values
X=d[:,0:4]
y=d[:,4]
colors=[color(y[i]) for i in range(y.shape[0])]
for i in range(4):
    for j in range(i+1,4):
        print(i,' ',j )
        plt.scatter(X[:,i], X[:,j],  c=colors)
        plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
#X = preprocessing.scale(X)
#X=100*X
pc=PCA(n_components=4)
pc.fit(X)
#X_PCA =pc.transform(X)
X_PCA =pc.fit_transform(X)
print(pc.components_)
print(pc.explained_variance_)
print(X[:,0].var(),X[:,1].var(),X[:,2].var(),X[:,3].var())
print(np.sum(pc.explained_variance_))
print(X[:,0].var()+X[:,1].var()+X[:,2].var()+X[:,3].var())
print(pc.explained_variance_ratio_)

for i in range(4):
    for j in range(i+1,4):
        print('PCA ',i,' ',j )
        plt.scatter(X_PCA[:,i], X_PCA[:,j],  c=colors)
        plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
XX = preprocessing.scale(X)
X_PCA = PCA(n_components=4).fit_transform(XX)
print(pc.components_)
print(pc.explained_variance_)
print(np.sum(pc.explained_variance_))
print(X[:,0].var(),X[:,1].var(),X[:,2].var(),X[:,3].var())
print(X[:,0].var()+X[:,1].var()+X[:,2].var()+X[:,3].var())
print(pc.explained_variance_ratio_)
for i in range(4):
    for j in range(i+1,4):
        print('PCA ',i,' ',j )
        plt.scatter(X_PCA[:,i], X_PCA[:,j],  c=colors)
        plt.show()
        
```

```python

from sklearn.neighbors import NearestNeighbors
neigh0 = NearestNeighbors(10)
neigh0.fit(X)
i=0
a=neigh0.kneighbors(X,10, return_distance=True)
print(a[0][i,:],a[1][i,:])

neigh1 = NearestNeighbors(10)
neigh1.fit(X_PCA)
a=neigh1.kneighbors(X_PCA, 10, return_distance=True)
print(a[0][i,:],a[1][i,:])
```

# PCA Details.

```python
XX = preprocessing.scale(X)
pca=PCA(n_components=4).fit(XX)
print('PC_components=', pca.components_)
print('variance_ratio=', pca.explained_variance_ratio_)
print('variance=', pca.explained_variance_)
```

SVD with Numpy


```python
XX = preprocessing.scale(X)
u, s, vh = np.linalg.svd(XX, full_matrices=True)
print('u=',u)
print('s=',s)
print('vh=',vh)
h=np.square(s)
h=h/np.sum(h)
print('s=',h)
```

t-SNE


```python
from sklearn.manifold import TSNE
XX = preprocessing.scale(X)
X_TSNE = TSNE(n_components=2).fit_transform(XX)
plt.scatter(X_TSNE[:,0], X_TSNE[:,1],  c=colors)
plt.show()
```

```python
from sklearn.manifold import TSNE
XX = preprocessing.scale(X)
X_TSNE = TSNE(n_components=2,perplexity=13,early_exaggeration=10).fit_transform(XX)
plt.scatter(X_TSNE[:,0], X_TSNE[:,1],  c=colors)
plt.show()
```
