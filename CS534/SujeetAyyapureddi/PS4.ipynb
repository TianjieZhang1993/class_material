```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
from sklearn.datasets import make_hastie_10_2
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt


def get_error_rate(pred, actual):
    return sum(pred != actual) / float(len(actual))

def logit_clf(Y_train, X_train, Y_test, X_test, clf):
    clf.fit(X_train,Y_train)
    pred_train = clf.predict(X_train)
    #print(pred_train)
    pred_test = clf.predict(X_test)
    return get_error_rate(pred_train, Y_train), \
           get_error_rate(pred_test, Y_test)


def adaboost_clf(Y_train, X_train, Y_test, X_test, M, clf):
    n_train, n_test = len(X_train), len(X_test)
    # Initialize weights
    w = np.ones(n_train) / n_train
    pred_train, pred_test = [np.zeros(n_train), np.zeros(n_test)]
    
    for i in range(M):
        # Fit a classifier with the specific weights
        clf.fit(X_train, Y_train, sample_weight = w)
        pred_train_i = clf.predict(X_train)
        pred_test_i = clf.predict(X_test)
        
        error = [int(x) for x in (pred_train_i != Y_train)]
        # Equivalent with 1/-1 to update weights
        error_sign = [x if x==1 else -1 for x in error]
        # Error Norm
        err_m = np.dot(w,error) / sum(w)
        # Alpha
        alpha_m = 0.5 * np.log( (1 - err_m) / float(err_m))
        # New weights
        w = np.multiply(w, np.exp([float(x) * alpha_m for x in error_sign]))
        # Add to prediction
        pred_train = [sum(x) for x in zip(pred_train, 
                                          [x * alpha_m for x in pred_train_i])]
        pred_test = [sum(x) for x in zip(pred_test, 
                                         [x * alpha_m for x in pred_test_i])]
        #print("pred_train", pred_train)
    
    pred_train, pred_test = np.sign(pred_train), np.sign(pred_test)
    # Return error rate in train and test set
    return get_error_rate(pred_train, Y_train), \
           get_error_rate(pred_test, Y_test)


def plot_error_rate(er_train, er_test, x_range):
    df_error = pd.DataFrame([er_train, er_test]).T
    df_error.columns = ['Training', 'Test']
    plot1 = df_error.plot(linewidth = 3, figsize = (8,6),
            color = ['lightblue', 'darkblue'], grid = True)
    plot1.set_xlabel('Number of iterations', fontsize = 12)
    plot1.set_xticklabels(x_range)
    plot1.set_ylabel('Error rate', fontsize = 12)
    plot1.set_title('Error rate vs number of iterations', fontsize = 16)
    plt.axhline(y=er_test[0], linewidth=1, color = 'red', ls = 'dashed')


if __name__ == '__main__':
    
    # Read data
    x, y = make_hastie_10_2() # taken from scikit
    #https://scikit-learn.org/stable/auto_examples/ensemble/plot_adaboost_hastie_10_2.html#sphx-glr-auto-examples-ensemble-plot-adaboost-hastie-10-2-py
    df = pd.DataFrame(x)
    df['Y'] = y

    # Split into training and test set
    train, test = train_test_split(df, test_size = 0.2)
    X_train, Y_train = train.iloc[:,:-1], train.iloc[:,-1]
    X_test, Y_test = test.iloc[:,:-1], test.iloc[:,-1]
    
    # Fit logistic regression
    clf = LogisticRegression(random_state=0, solver='lbfgs')
    er_logit = logit_clf(Y_train, X_train, Y_test, X_test, clf)
    
    
    er_train, er_test = [], []
    #er_train, er_test = [er_logit[0]], [er_logit[1]]
    
    #print(er_train, er_test)
    er_train.append(er_logit[0])
    er_test.append(er_logit[1])
    #print(er_train, er_test)
    x_range = range(1, 20, 1)
    for i in x_range:    
        er_i = adaboost_clf(Y_train, X_train, Y_test, X_test, i, clf)
        er_train.append(er_i[0])
        er_test.append(er_i[1])
        #print(er_train, er_test)
    
    # Compare error rate vs number of iterations
    plot_error_rate(er_train, er_test,x_range)
```

```python
from sklearn.cluster import DBSCAN

data = pd.read_csv("iris.csv",header=None)
data.columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'y']

def convert_string_to_color(x):
    if x== 'Iris-setosa':
        return 0
    elif x== 'Iris-versicolor':
        return 1
    else:
        return 2
    

data["color"] = data['y'].apply(convert_string_to_color)
data_val=data.values


X=data_val[:,0:4]
y=data_val[:,4]

scaler = StandardScaler()
X_S = scaler.fit_transform(X)

```

```python
array_eps= np.arange(0.1, 1.4, 0.1)
dict_eps= {}
for ep in array_eps:
    db = DBSCAN(eps=ep, min_samples=10).fit(X_S)
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_

    # Number of clusters in labels, ignoring noise if present.
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise_ = list(labels).count(-1)
    dict_eps[ep] = n_clusters

```

```python
lists = sorted(dict_eps.items())
x, y = zip(*lists)

#plotting 
plt.plot(x, y,color='r', marker='o')
plt.xlabel("eps Values")
plt.ylabel("Number of Clusters")
plt.show()
```
