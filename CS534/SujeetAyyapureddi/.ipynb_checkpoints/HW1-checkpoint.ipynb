### Q2

###  The sum of the squares of the columns along the row direction of the PCA_ matrix results in 1. The correlation between a component and a variable is a measure of the information common to both.  As each component is correlated to each variable,  the variable which has common information with a component will have a higher correlation coefficient.

### As the product of the Principal components and the variable  leads to the trasformed data, the coefficient of the principal component corresponding to the highly correlated variable will have a higher value(abs value, as the square of  the coefficients matters).

### This is the observation.

### one more thing to note is that the row in the output of the PCA_ are the principal coeffecients. The output of the PCA_ needs to be transposed and multiplied with the scaled data to get the transformed data.

```python
length = df.shape[0]
dependent_class = df.columns[-1]
dependent_class,length
```

```python
feat =df.columns[0:len(df.columns)-1]
feat_list = feat.tolist()
```

```python
def gen_gini_index(dframe):
    prob_square =0
    classes = dframe[dependent_class].unique()
    length = dframe.shape[0]
    for variables in classes:
        prob = dframe[dependent_class].value_counts()[variables]/length
        prob_square += np.power(prob,2)
    
    return(1 - prob_square)
    
def gen_entropy(dframe):
    entropy =0
    classes = dframe[dependent_class].unique()
    length = dframe.shape[0]
    for variables in classes:
        prob = dframe[dependent_class].value_counts()[variables]/length
        entropy += -1*np.log(prob)*prob
    
    return entropy  
    
    
    
```

```python
def gen_gini_index_weighted(dframe, independent_feature):
    wt_gini =0
    #classes = dframe[dependent_class].unique()
    # variable is the values the independent feature can take 
    for variable in dframe[independent_feature].unique():
        part_var_df=dframe.loc[dframe[independent_feature]==variable]
        wt_gini+=(part_var_df.shape[0]/length)*gen_gini_index(part_var_df)
        
    return wt_gini 

def gen_gain_ratio(dframe, independent_feature):
    sigma_entropy =0
    split_info =0
    #classes = dframe[dependent_class].unique()
    # variable is the values the independent feature can take 
    for variable in dframe[independent_feature].unique():
        part_var_df=dframe.loc[dframe[independent_feature]==variable]
        split_info+=-1*(part_var_df.shape[0]/length)*np.log2((part_var_df.shape[0]/length))
        sigma_entropy+=(part_var_df.shape[0]/length)*gen_entropy(part_var_df)
     
    gain_ratio = (gen_entropy(dframe) - sigma_entropy)/split_info
    return gain_ratio 
    
```

```python
# Returns True when splitting needs to stop
def stop_condition(dframe):
    class_count = dframe[dependent_class].value_counts()
    
    if class_count.shape[0] > 1 :
        return False
    else:
        return True

# Stop the spliting if the split results in a larger ginni index. This implies that the stop condition should kick in.
# Returns True when splitting needs to stop
def stop_condition_ginni(dframe):
    
    base_gini_index = gen_gini_index(dframe)
    
    
    #create list of features
    feat =dframe.columns[0:len(dframe.columns)-1]
    feat_list = feat.tolist()
    
    
    
    for feature in feat_list:
        wt_gini =0
        for variable in dframe[feature].unique():
            
            part_var_df=dframe.loc[dframe[feature]==variable]
            wt_gini+=(part_var_df.shape[0]/length)*gen_gini_index(part_var_df)
            
        #print(feature, wt_gini)
        
        if base_gini_index >= wt_gini: # return False if a certain split is helpful
            return False
         
    return True


def create_node(dframe):
    variable = dframe[dependent_class].value_counts().idxmax()
    return variable
 
def gen_best_split_feature(dframe):
    
    gin_index=gen_gini_index(dframe)
    
    
    #create list of features
    feat =dframe.columns[0:len(dframe.columns)-1]
    feat_list = feat.tolist()
    split_feature =""
    
    for feature in feat_list:
        ginni_idx=gen_gini_index_weighted(dframe, feature)
        if ginni_idx < gin_index:
            gin_index = ginni_idx
            split_feature = feature
            
    return split_feature
    
def gen_best_split_feature_gain_ratio(dframe):
    
    gin_index=gen_gini_index(dframe)
    gain_ratio_min = 0
    
    #create list of features
    feat =dframe.columns[0:len(dframe.columns)-1]
    feat_list = feat.tolist()
    split_feature =""
    
    for feature in feat_list:
        gain_ratio=gen_gain_ratio(dframe, feature)
        if gain_ratio > gain_ratio_min:
            gain_ratio_min = gain_ratio
            split_feature = feature
            
    return split_feature
```

```python
def gen_tree(dframe, trace_list):
    temp_list=[]
    
    
    #base case. Stop splitting if the base case is triigered
    if(stop_condition(dframe) or (stop_condition_ginni(dframe)) or (dframe.shape[1] == 1)):
        temp_list.append(create_node(dframe))
        trace_list.append(temp_list)
        return

    temp_df=dframe
    split_feat=gen_best_split_feature(temp_df)
    
    
    for feat_value in dframe[split_feat].unique():
        temp_list=[]
        
        temp_df=dframe
        #Append information from this loop into the local list
        temp_list.append(split_feat)
        temp_list.append(feat_value)
        #Append information from this loop into the global list
        trace_list.append(temp_list)
        next_level_df = temp_df.loc[temp_df[gen_best_split_feature(temp_df)]==feat_value].drop(columns=gen_best_split_feature(temp_df))
        
        gen_tree(next_level_df, trace_list)
        
    return


def gen_tree_gain_ratio(dframe, trace_list):
    temp_list=[]
    
    
    #base case. Stop splitting if the base case is triigered
    if(stop_condition(dframe) or (dframe.shape[1] == 1)):
        temp_list.append(create_node(dframe))
        trace_list.append(temp_list)
        return

    temp_df=dframe
    split_feat=gen_best_split_feature_gain_ratio(temp_df)
    
    
    for feat_value in dframe[split_feat].unique():
        temp_list=[]
        
        temp_df=dframe
        
        temp_list.append(split_feat)
        temp_list.append(feat_value)
        
        trace_list.append(temp_list)
        next_level_df = temp_df.loc[temp_df[gen_best_split_feature_gain_ratio(temp_df)]==feat_value].drop(columns=gen_best_split_feature_gain_ratio(temp_df))
        
        gen_tree_gain_ratio(next_level_df, trace_list)
        
    return
```

```python
trace_list=[]
gen_tree(df, trace_list)
```
